## Dot Product

```python
np.random.seed(0)

mat1 = np.random.randint(10, size=(5,3))
mat2 = np.random.randint(10, size=(5,3))

mat1, mat2
```

```python
# element-wise multiplication (Hadamard product)
mat1 * mat2
```

```python
# dot product

mat3 = np.dot(mat1, mat2.T)
```

```python
mat3.shape
```

## Dot product example, Nut butter sales

```python
np.random.seed(0)
# Number of jars sold
sales_amount = np.random.randint(20, size=(5,3))
sales_amount
```

```python
# Create weekly sales DataFrame  
weekly_sales = pd.DataFrame(sales_amount,
                           index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
                           columns=['Almond Butter', 'Peanut Butter', 'Cashew Butter'])
weekly_sales
```

```python
# Create prices array
prices = np.array([10,8,12])
prices
```

```python
prices_data_frame = pd.DataFrame(prices.reshape(1,3),
                                index=['Price'],
                                columns=['Almond Butter', 'Peanut Butter', 'Cashew Butter'])
prices_data_frame
```

```python
total_sales = prices.dot(sales_amount.T)
total_sales
```

```python
# create daily sales
```

```python
prices_data_frame
```

```python
weekly_sales
```

```python
prices_data_frame.shape, weekly_sales.shape
```

```python
daily_sales = prices_data_frame.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales['Total ($)'] = daily_sales.T
weekly_sales
```
